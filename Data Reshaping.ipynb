```r
stocks = data.frame( 
            time = as.Date('2009-01-01') + 0:9,
            X = rnorm(10,0,1),
            Y = rnorm(10,0,2),
            Z = rnorm(10,0,4))

stocks
```

```r
stocks.gathered = stocks %>% gather(stock, price, X:Z)
stocks.gathered
```

```r
spread(stocks.gathered, stock, price)
```

```r
df = data.frame(new.col = c(NA, "a.x", "b.y", "c.z"))
df
```

```r
df = data.frame(new.col = c(NA, "a-x", "b-y", "c-z"))
df
```
